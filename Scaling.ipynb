```python
from sklearn import datasets as dst
import pandas as pd
```

```python
boston = dst.load_boston()
```

```python
df = pd.DataFrame(boston.data)
df.describe()
```

```python
from sklearn import preprocessing  as ps#for scalaing we need to import this module 
```

```python
scaled_data = ps.scale(boston.data)
df2 = pd.DataFrame(scaled_data)
df2.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test, y_train, y_test = train_test_split(scaled_data,boston.target,test_size=0.2)
```

### This scale function makes mean zero i.e subtracts mean from all

#### while scaling we shouldn't consider test data so we should first split .

```python
from sklearn.preprocessing import StandardScaler as sds
```

```python
scaler = sds()
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```
