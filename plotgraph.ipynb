```python
import matplotlib.pyplot as plt
```

```python
x = [5,8,10]
y = [10,16,20]
plt.grid(True)
plt.plot(x,y,color = 'g',label = "val 1")

x1 = [15,18,20]
y1 = [16,18,20]
plt.plot(x1,y1, 'ro',label="val2") # ro is a marker , if r^ will be written it will give triangle shaped  marker in plot
#if we specify plot and marker in same plot we won't need to use scatter 

x2 = [2,3,5]
y2 = [1,4,7]
plt.plot(x2,y2,color='r',marker='*',label="val3")
plt.scatter(x,y,color = 'c')
plt.text(4,9 , "ab" )
plt.legend() #used after using label 
plt.xlabel("X")
plt.ylabel("Y")
plt.title("DemoGraph")
plt.show()
```

```python
plt.bar(x,y,color='m')
plt.grid(True)
plt.show()
```
